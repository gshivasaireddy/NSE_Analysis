```python
import stock_object 
import stock_price_extraction as spe
import balance_sheet_extraction as ble 
import income_statement_extraction as ise
import share_holders_extraction as she
import cashflow_extraction as ce
import warnings
import pandas as pd
warnings.filterwarnings('ignore')
```

```python
nse_company_file = 'C:\\Users\\shiva.reddy\\Personal_Projects\\NSE_Analysis\\NSE_Analysis\\NSE_Project\Input\\NSE_Listed_Companies_SectorWise_Sample.csv'
df_nse = pd.read_csv(nse_company_file,sep='~')



# Consolidation DataFrames declaration
df_final_stock_prices_consolidation = pd.DataFrame()

df_final_bs_yearly_consolidation = pd.DataFrame()
df_final_bs_quarterly_consolidation = pd.DataFrame()

df_final_is_yearly_consolidation = pd.DataFrame()
df_final_is_quarterly_consolidation = pd.DataFrame()

df_final_cf_yearly_consolidation = pd.DataFrame()
df_final_cf_quarterly_consolidation = pd.DataFrame()

for index ,row in df_nse.iterrows():
    print(row)
    print('Company : ',row['Ticker'])
    stock = stock_object.Stock(row['nse_ticker'])
    stock_obj = stock.getStockObject()

    # price history
    print('Price History: ')
    stock_price_obj = spe.StockPriceExtraction(stock_obj , row['Ticker'])
    df_stock_price_company_wise = stock_price_obj.getPriceHistory()
    df_final_stock_prices_consolidation = pd.concat([df_final_stock_prices_consolidation,df_stock_price_company_wise],ignore_index=True)
    
    # balance sheet
    balance_sheet_obj =  ble.BalanceSheetExtraction(stock_obj,row['Ticker'])
    df_balance_sheet_quartely = balance_sheet_obj.getBalanceSheetQuarterly()
    df_balance_sheet_yearly =  balance_sheet_obj.getBalanceSheetYearly()
    df_final_bs_yearly_consolidation= pd.concat([df_final_bs_yearly_consolidation,df_balance_sheet_yearly])
    df_final_bs_quarterly_consolidation = pd.concat([df_final_bs_quarterly_consolidation,df_balance_sheet_quartely])



    # income statement
    income_stmt_obj  = ise.IncomeStatementExtraction(stock_obj,row['Ticker'])
    df_income_stmt_yearly = income_stmt_obj.getIncomeStatementYearly()
    df_income_stmt_quarterly = income_stmt_obj.getIncomeStatementQuarterly()
    df_final_is_yearly_consolidation=pd.concat([df_final_is_yearly_consolidation,df_income_stmt_yearly])
    df_final_is_quarterly_consolidation=pd.concat([df_final_is_quarterly_consolidation,df_income_stmt_quarterly])




    # cashflow
    cashflow_obj =  ce.CashFlowExtraction(stock_obj,row['Ticker'])
    df_cashflow_yearly =  cashflow_obj.getCashFlowYearly()
    df_cashflow_quarterly =  cashflow_obj.getCashFlowQuarterly()
    df_final_cf_yearly_consolidation=pd.concat([df_final_cf_yearly_consolidation,df_cashflow_yearly])
    df_final_cf_quarterly_consolidation=pd.concat([df_final_cf_quarterly_consolidation,df_cashflow_quarterly])



    # All Listed companies- Day level Price information of the stock consolidation
    # df_final_stock_prices = pd.concat([df_final_stock_prices,df_stock_price_company_wise],ignore_index=True)
    
    
```
